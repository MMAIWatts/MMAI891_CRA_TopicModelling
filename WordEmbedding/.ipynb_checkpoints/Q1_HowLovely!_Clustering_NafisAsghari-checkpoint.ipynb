## Standardization

```python
X = jewelry_customers.copy()

sc = StandardScaler()
X = sc.fit_transform(X)
X.shape
```

# k-Means

## Find the Best k

```python
inertias = {}
silhouettes = {}
for k in range(2, 11):
    kmeans = KMeans(init='k-means++', n_init=10, n_clusters=k, max_iter=1000, random_state=42).fit(X)
    inertias[k] = kmeans.inertia_ # Inertia: Sum of distances of samples to their closest cluster center
    silhouettes[k] = silhouette_score(X, kmeans.labels_, metric='euclidean')
    

plt.figure()
plt.grid(True);
plt.plot(list(inertias.keys()), list(inertias.values()))
plt.title('K-Means, Elbow Method')
plt.xlabel("Number of clusters, K")
plt.ylabel("Inertia")


plt.figure()
plt.grid(True)
plt.plot(list(silhouettes.keys()), list(silhouettes.values()))
plt.title('K-Means, Elbow Method')
plt.xlabel("Number of clusters, K")
plt.ylabel("Silhouette")
plt.show()
```

- The first plot shows the value of the Within-Cluster-Sum of Squared Errors (WCSS) for different values of k. At k = 5, improvement in distortion (WCSS) declines the most (the Elbow). So based on this method, k = 5 is the optimal number for clusters.

- The second plot illustrates average silhouette values for different values of k. At k = 5, this value is maximum, which indicates that most records are well matched to their own clusters and poorly matched to neighbouring clusters.


__Therefore, k = 5 is the optimal value for the number of clusters based on both metrics.__

## K-means with Optimal Number of Clusters

```python
#k-means with 5 clusters
kmeans = KMeans(init='k-means++', n_clusters=5, random_state=42).fit(X)
centroid = kmeans.cluster_centers_
centroid
```

```python
centroid_true_values = sc.inverse_transform(centroid)

pd.DataFrame(centroid_true_values , columns = jewelry_customers.columns)
```

```python
from yellowbrick.cluster import SilhouetteVisualizer

visualizer = SilhouetteVisualizer(kmeans)
visualizer.fit(X)
visualizer.poof()
fig = visualizer.ax.get_figure()
```

```python
labels = kmeans.labels_
unique, counts = np.unique(labels, return_counts=True) #count the occurrences of each label
dict(zip(unique, counts))
```

```python
print(f"WCSS: {kmeans.inertia_:.2f}")
print(f"silhouette_score: {silhouette_score(X, kmeans.labels_):.2f}")
```

```python
#use true values (not normalize values) to understand each cluster better
for label in set(kmeans.labels_):
    print('\nCluster {}:'.format(label))
    X_tmp = jewelry_customers[kmeans.labels_==label].copy()
    X_tmp.loc['mean'] = X_tmp.mean()
    X_tmp.head(10)
```

```python
#Find Examples

X = pd.DataFrame(X, columns = jewelry_customers.columns)
samples = {}
from scipy.spatial import distance

for i, label in enumerate(set(kmeans.labels_)):    
    X_tmp = X[kmeans.labels_==label].copy()
    
    exemplar_idx = distance.cdist([kmeans.cluster_centers_[i]], X_tmp).argmin()
    df_t = jewelry_customers[kmeans.labels_==label].copy()
    exemplar = pd.DataFrame(df_t.iloc[exemplar_idx])
    samples[f"Cluster {i}"] = exemplar
    print('\nCluster {}:'.format(label))
    exemplar
```

```python
s = pd.concat([samples['Cluster 0'], samples['Cluster 1'],samples['Cluster 2'],
           samples['Cluster 3'], samples['Cluster 4']], axis = 1)
s.columns = ["Cluster_0", "Cluster_1","Cluster_2","Cluster_3","Cluster_4"]

s.round(2)
```

# Hierarchical Clustering

```python
# Using the dendrogram to find the optimal number of clusters
import scipy.cluster.hierarchy as sch
sns.set(style = "white")
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

n_clusters = 5 seems to be a reasonable choice.

```python
# Fitting Hierarchical Clustering to the dataset
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```

```python
unique, counts = np.unique(hc.labels_, return_counts=True) #count the occurrences of each label
dict(zip(unique, counts))
```

# DBSCAN

```python
silhouettes = {}
for eps in np.arange(0.1, 0.6, 0.1):
    db = DBSCAN(eps=eps, min_samples=3).fit(X)
    silhouettes[eps] = silhouette_score(X, db.labels_, metric='euclidean')
    

plt.figure()
plt.plot(list(silhouettes.keys()), list(silhouettes.values()))
plt.title('DBSCAN, Elbow Method')
plt.xlabel("Eps");
plt.ylabel("Silhouette")
plt.grid(True)
```

At eps = 0.3 we can see an elbow. Therefore, we set eps parameter to 0.3

```python
db = DBSCAN(eps=0.3, min_samples=2).fit(X)
```

```python
#db.labels_
unique, counts = np.unique(db.labels_, return_counts=True)
dict(zip(unique, counts))
# -1 means no cluster is assigned to that customer
```
