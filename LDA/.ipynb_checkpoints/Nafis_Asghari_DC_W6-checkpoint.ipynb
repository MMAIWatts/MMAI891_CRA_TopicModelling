## Cuisine with most and least Critical Ratio

```python
cs_flags = df[['CUISINE DESCRIPTION', 'CRITICAL FLAG','ACTION' ]].groupby(['CUISINE DESCRIPTION','CRITICAL FLAG']).count()
```

```python
cs_flags['prect'] = (cs_flags.ACTION / cs_flags.groupby(level=0).ACTION.transform(sum) * 100)
```

```python
cs_flags.head()
```

```python
cs_filtered = cs_flags[np.in1d(cs_flags.index.get_level_values(1), ['Critical'])]
cs_filtered = cs_filtered.sort_values(by=['prect'])
```

```python
# Cuisine with least critical ratio
round(cs_filtered[['prect']].head(15), 2)
```

```python
# Cuisine with most critical ratio
round(cs_filtered[['prect']].tail(15).iloc[::-1], 2)
```
